### 4. Penyelarasan kurva pertumbuhan seluruh negara dengan jumlah kasus kematian COVID-19 â‰¥ 25 (Bobot: 25%)

**Untuk mendapatkan gambaran perkembangan kasus covid-19 di berbagai negara, diperlukan penyelarasan kurva pertumbuhan setiap negara. Kurva hanya menampilkan informasi yang dimulai dengan hari dimana data kasus kematian COVID-19 setidaknya 25 orang.**

```python
#type your codes here
dftemp_death25 = raw_data_deaths.copy()
dftemp_death25
del dftemp_death25["Province/State"]
del dftemp_death25["Lat"]
del dftemp_death25["Long"]
dfPerCountry_death25 = dftemp_death25.groupby(["Country/Region"]).sum()
dfPerCountry_death25

df = None
for i in dfPerCountry_death25.columns:
    negara  = dfPerCountry_death25.index.tolist()
    tanggal = [i]*len(negara)
    jumlah  = dfPerCountry_death25[i].tolist()
    raw_data = {'Country/Region': negara,'Date': tanggal, 'Total Kematian':jumlah}
    dftgl = pd.DataFrame(raw_data)
    df = pd.concat([df,dftgl],axis = 0)
dfnegaratgl = df.groupby(["Country/Region","Date"]).sum()
dfplot = dfnegaratgl[dfnegaratgl["Total Kematian"]>=25]
dfplot
```

##### Tuliskan analisis anda disini:
Dari data frame di atas, kita dapat melihat pada tanggal berapa negara tersebut yang jumlah kematiannya minimal 25 orang dalam 1 hari, jika ada negara yang tidak mencapai 25 maka negara tersebut tidak ditampilkan.

**Visualisasikan hasil diatas dan berikan/atur judul, labels, dan spesifikasi (ukuran, warna, ketebalan, dll) yang sesuai sehingga plot/kurva yang dihasilkan rapi, menarik, dan mudah dipahami.**

```python
#plot time series
# type your codes here
listnegara = set(df['Country/Region'].tolist())
listnegara
for namanegara in listnegara:
    dfpernegara = df[(df['Country/Region']==namanegara) & (df['Total Kematian']>25)][["Date",'Total Kematian']]
    if len(dfpernegara)>0:
        dfpernegara.set_index("Date").plot()
        plt.title("Kasus Kematian COVID-19 Negara " + namanegara )
        plt.xlabel("Tanggal")
        plt.ylabel("Jumlah Kematian")
```

##### Tuliskan analisis anda disini:
Kita dapat melihat visualisasi dari negara-negara yang kasus kematian akibat COVID-19 minimal 25 dan rata-rata tiap negara tersebut mengalami kenaikan sampai tanggal 4/7/20.

### 5. Visualisasi jumlah pasien yang sembuh (recovered people) (Bobot: 25%)
    
**Anda perlu mendapatkan informasi mengenai total pasien COVID-19 yang dinyatakan sembuh seperti pada kasus terkonfirmasi dan kasus kematian tapi, kali ini lakukan hal tersebut dengan membuat `user-defined functions` yang diperlukan bukan mengetik ulang barisan code yang sama seperti di nomor sebelumnya.**

```python
#type your function codes here
def cleaning_data():
    dftemp_recovered = raw_data_recovered.copy()
    del dftemp_recovered["Province/State"]
    del dftemp_recovered["Lat"]
    del dftemp_recovered["Long"]
    dfPerCountry_recovered = dftemp_recovered.groupby(["Country/Region"]).sum() 
    return dfPerCountry_recovered  

def calculate_for_all_date(dfpcr):
    dfsum = dfpcr.sum(axis=1).sort_values(ascending=False)
    return dfsum

def calculate_total(dfsum):
    return dfsum.sum()

def display_summary(dfdetail, total):
    print("LAPORAN TOTAL SEMBUH PER NEGARA")
    for index, val in dfdetail.iteritems():
        if val>0:
            print(index," = ", val)
    print("\nLAPORAN NEGARA YANG BELUM ADA YANG SEMBUH")
    for index,val in dfdetail.iteritems():
        if val==0:
            print(index," = ", val)
    print()
    print("Total Sembuh semuanya : ", total)
```

**Periksa fungsi-fungsi yang sudah anda buat**

```python
# write your codes to check your functions here
dfpcr = cleaning_data()
dfsum = calculate_for_all_date(dfpcr)
total = calculate_total(dfsum)
display_summary(dfsum,total)
```

**Selanjutnya,tampilkan data total pasien COVID-19 yang sembuh dalam bentuk grafik yang sesuai. Berikan judul, labels, dan spesifikasi (ukuran, warna, ketebalan, dll) yang sesuai sehingga plot yang dihasilkan rapi, menarik, dan mudah dipahami.**

```python
#type your codes here
dftemp_recovered = raw_data_recovered.copy()
del dftemp_recovered["Province/State"]
del dftemp_recovered["Lat"]
del dftemp_recovered["Long"]
del dftemp_recovered["Country/Region"]

dftemp_recovered.sum().plot()
plt.title("Total Pasien COVID-19 Yang Sembuh")
plt.xlabel("Tanggal")
plt.ylabel("Jumlah Sembuh")
```
